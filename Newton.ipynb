```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pylab
```

Изобразим функции $y_1 = \cos(x)$ и $y_2 = \frac{1}{x-3}$ дабы локализовать корни.

```python
x = np.linspace(-4*3.1415, 4*3.1415, 100)
y1 = np.cos(x)
y2 = []
for i in range(len(x)):
    y2.append(1 / (x[i] - 3))
plt.figure(figsize=(8, 5.5))
plt.plot(x, y1)
plt.plot(x, y2)
plt.ylim([-13, 6])
plt.xlim([-13, 13])

plt.grid()
pylab.xticks(range(-13, 13))
pylab.yticks(range(-13, 7))
plt.show()
```

$\newline$
Легко видеть, что это действительно так.

Изобразим график $f(x) = -\arccos(\frac{1}{x-3})$, чтобы убедиться, что $f([-2, -1]) \subseteq [-2, -1]$.

```python
x1 = np.linspace(-7, 1.99, 160)
x2 = np.linspace(4.0001, 7, 70)
x = np.hstack((x1, x2))
y1 = np.zeros(230) 
y21 = []
y22 = []
for i in range(len(x1)):
    y21.append(-math.acos(1 / (x1[i] - 3)))
for i in range(len(x2)):
    y22.append(-math.acos(1 / (x2[i] - 3)))

plt.figure(figsize=(8, 5.5))
plt.plot(x, y1, color = 'black')
plt.plot(y1, x, color = 'black')
plt.plot(x1, y21, color = 'orange')
plt.plot(x2, y22, color = 'orange')
plt.ylim([-4, 1])
plt.xlim([-3, 6])

plt.grid()
plt.show()
```

Убеждаемся в том, что график функции лежит в квадрате $[-2, -1]\times[-2, -1]$, т.е. оба уcловия сходимости выполнены.

```python
def FPI(x):
    k = 1
    x_n1 = 0
    x_n = x
    x_1 = -math.acos(1/(x_n - 3))
    while (abs(x_1 - x_n) > (10**(-3))):
        x_n1 = x_n
        x_n = x_1
        x_1 = -math.acos(1/(x_1 - 3))
        k += 1
    return x_1, k
```

```python
def Newton(x, a, b):
    x_old = 0
    k = 0
    while (abs(x - x_old) > 10**(-4)):
        x_old = x
        x = x - ((x - 3)*math.cos(x) - 1) / ((2*x-3)*math.cos(x) - (x**2 - 3*x)*math.sin(x))
        while (True):
            if ((x > a) and (x < b)):
                break
            else :
                x = (x_old + x) / 2
        k += 1
    return x, k
```
